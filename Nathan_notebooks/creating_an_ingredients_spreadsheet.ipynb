```python
import pandas as pd
```

```python
df = pd.read_pickle("../data/rec_ingredients.pk")
df
```

```python
ingredients_list = []
from collections import defaultdict 
ingredient_popularity_dictionary = defaultdict(lambda : 0)
for entry_list in list(df.RecipeIngredientParts):
    ingredients_list += entry_list
    for entry in entry_list:
        ingredient_popularity_dictionary[entry] += 1

ingredients_list = sorted(list(set(ingredients_list)), key=lambda item: ingredient_popularity_dictionary[item], reverse=True)
```

```python
ingredient_df = pd.DataFrame({"original": ingredients_list, "simplified": ingredients_list})
ingredient_df["Popularity"]=ingredient_df.original.map(ingredient_popularity_dictionary)
ingredient_df
```

```python
# Read in work I've already done to manually correct the first 430 ingredients
previously_simplified_ingredients_df = pd.read_excel("ingredient_data/manually_generated_ingredient_simplification_spreadsheet.xlsx")
previous_simplification_dict = pd.Series(previously_simplified_ingredients_df.iloc[:430,:]["simplified"].values,index=previously_simplified_ingredients_df.iloc[:430,:]["original"]).to_dict()
# previously_simplified_ingredients_df.iloc[:430,:].to_dict()
previous_simplification_dict
```

```python
ingredient_df["simplified"] = ingredient_df["original"].apply(lambda ingredient: previous_simplification_dict.get(ingredient, ingredient))
ingredient_df.to_excel("ingredient_data/manually_generated_ingredient_simplification_spreadsheet_sorted_by_popularity.xlsx", index=False)
ingredient_df
```
