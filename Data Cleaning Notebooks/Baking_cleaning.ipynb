```python
# Loading packages
import pandas as pd
import os
```

```python
# Loading Keywords
keywords_path = '../data/clean_columns/keywords_clean.parquet'
keywords_df = pd.read_parquet(keywords_path)
```

## Creating the `baking` Column

```python
# Cooking related keywords
baking_keywords = ['breads', 'cookie & brownie', 'pie', 'bar cookie', 'drop cookies', 'yeast breads', 'cheesecake']

# Create the `baking` column
def is_baking(row):
    for keyword in row['KeywordsClean']:
        if any(baking_keyword in keyword.lower() for baking_keyword in baking_keywords):
            return True
    return False

keywords_df['baking'] = keywords_df.apply(is_baking, axis=1)
keywords_df.head()
```

## Saving Cleaned Data

```python
# Save DataFrame containing the cleared `baking` column
keywords_df[['Name', 'baking']].to_parquet('../data/clean_columns/baking.parquet')
```
