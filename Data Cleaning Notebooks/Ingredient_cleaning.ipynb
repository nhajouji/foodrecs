```python
import pandas as pd
import numpy as np
```

```python
rec_ingredients_df = pd.read_pickle('../data/rec_ingredients.pk')
ingredientsimptable = pd.read_csv('../data/ingredientsimplifications.csv')
```

```python
ingredientsimptable.set_index('original',inplace=True)
```

```python
simplified_ingredients_list = []
for w in ingredientsimptable.simplified:
    simplified_ingredients_list+=w.split(', ')

simplified_ingredients_list = list(set(simplified_ingredients_list))
```

```python
ingredient_sim_dict = {ingredient:ingredientsimptable['simplified'][ingredient] for ingredient in ingredientsimptable.index}
```

```python
rec_ingredients_df
```

```python
simplified_ingredients_all = []
unsimplified_ingredients_all = []

for recingreds in rec_ingredients_df.RecipeIngredientParts:
    simplifiable = [ing for ing in recingreds if ing in ingredient_sim_dict]
    unsimplified = [ing for ing in recingreds if ing not in simplifiable]
    simplified = [ingredient_sim_dict[ing] for ing in simplifiable]
    simplified_ingredients_all.append(simplified)
    unsimplified_ingredients_all.append(unsimplified)

rec_ingredients_df['Simplified'] = simplified_ingredients_all
rec_ingredients_df['Unsimplified'] = unsimplified_ingredients_all
```

```python
rec_ingredients_df.loc[rec_ingredients_df['Unsimplified'].apply(len)>0]
```

```python
unclassified = []
for inglist in rec_ingredients_df.Unsimplified:
    unclassified+=inglist

unclassified_counts = {ing:0 for ing in set(unclassified)}
for ing in unclassified:
    unclassified_counts[ing]+=1

unclassified = list(set(unclassified))

unclassified.sort(key = lambda x : unclassified_counts[x],reverse = True)
```

```python
ingredient_sim_dict.update({'Quinoa-Toasted':'quinoa','cannellini':'pasta','oleo':'butter','green pea':'green pea','frozen green pea':'green pea','jelly beans':'jelly beans'})
```

```python
bacon_ings = [w for w in unclassified if 'bacon' in w]
bacon_veg_ings = [w for w in bacon_ings if 'veg' in w]
bacon_nonveg_ings = [w for w in bacon_ings if w not in bacon_veg_ings]
```

```python
for w in bacon_veg_ings:
    ingredient_sim_dict[w] = 'vegetarian bacon'

for w in bacon_nonveg_ings:
    ingredient_sim_dict[w] = 'bacon'
```

```python
jellies = [w for w in unclassified if 'jell' in w and 'bean' not in w]
for w in jellies:
    ingredient_sim_dict[w] = 'jelly'
```

```python
rice_words = [w for w in unclassified if 'rice' in w and 'vinegar' not in w and 'flour' not in w and 'noodle' not in w and 'paper' not in w and 'mix' not in w]  
for w in rice_words:
    ingredient_sim_dict[w] = 'rice'
```

```python
ingredient_subwords_unc = []
for w in unclassified:
    ingredient_subwords_unc+=w.split(' ')

subword_counts = {w:0 for w in set(ingredient_subwords_unc)}
for w in ingredient_subwords_unc:
    subword_counts[w]+=1

subwords = list(subword_counts.keys())
subwords.sort(key = lambda x:subword_counts[x],reverse=True)
```
