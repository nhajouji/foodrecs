```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cooktimes = pd.read_pickle('../data/rectimes.pk')
```

```python
cooktimes.head()
```

```python
cooktimes=cooktimes.fillna('0S')
```

```python
def clean_cooktime(cooktime:str)->float:
    if type(cooktime)!=str:
        return cooktime
    elif cooktime[-1]=='S':
        return 0
    time = 0
    digits = [str(n) for n in range(10)]
    while cooktime[0] not in digits:
        cooktime = cooktime[1:]
    while len(cooktime)>0:
        i = 0
        while cooktime[i] in digits:
            i+=1
        tn = int(cooktime[:i])
        if cooktime[i] == 'H':
            time+=60*tn
            cooktime = cooktime[i+1:]
        else:
            time+=tn
            return time
    return time
```

```python
cooktimes['CookTimeClean']=[clean_cooktime(s) for s in cooktimes.CookTime]
cooktimes['PrepTimeClean']=[clean_cooktime(s) for s in cooktimes.PrepTime]
cooktimes['TotalTimeClean']=[clean_cooktime(s) for s in cooktimes.TotalTime]
```
